```python
# Step 1: Install Required Libraries
# Run this cell to install the required libraries if not already installed
!pip install pandas textblob

# Step 2: Import Libraries
import pandas as pd
from textblob import TextBlob

# Step 3: Load the CSV File
# Replace 'your_file.csv' with the name of your CSV file
file_path = "C:\\Users\\Shree\\Downloads\\SA1.csv"
df = pd.read_csv(file_path)

# Step 4: Define a Function for Sentiment Analysis
def analyze_sentiment(comment):
    analysis = TextBlob(comment)
    if analysis.sentiment.polarity > 0:
        return "Positive"
    elif analysis.sentiment.polarity < 0:
        return "Negative"
    else:
        return "Neutral"

# Step 5: Perform Sentiment Analysis on the 3rd Column
# Assuming the 3rd column is index 2 (0-based indexing)
df['Emotion'] = df.iloc[:, 2].apply(analyze_sentiment)

# Step 6: Save the Results to a New CSV File
output_file = "sentiment_analysis_results.csv"
df.to_csv(output_file, index=False)

# Display the first few rows of the updated DataFrame
df.head()
```
